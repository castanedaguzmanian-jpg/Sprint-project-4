- `instacart_orders.csv`: cada fila corresponde a un pedido en la aplicaci√≥n Instacart.
    - `'order_id'`: n√∫mero de ID que identifica de manera √∫nica cada pedido.
    - `'user_id'`: n√∫mero de ID que identifica de manera √∫nica la cuenta de cada cliente.
    - `'order_number'`: el n√∫mero de veces que este cliente ha hecho un pedido.
    - `'order_dow'`: d√≠a de la semana en que se hizo el pedido (0 si es domingo).
    - `'order_hour_of_day'`: hora del d√≠a en que se hizo el pedido.
    - `'days_since_prior_order'`: n√∫mero de d√≠as transcurridos desde que este cliente hizo su pedido anterior.
- `products.csv`: cada fila corresponde a un producto √∫nico que pueden comprar los clientes.
    - `'product_id'`: n√∫mero ID que identifica de manera √∫nica cada producto.
    - `'product_name'`: nombre del producto.
    - `'aisle_id'`: n√∫mero ID que identifica de manera √∫nica cada categor√≠a de pasillo de v√≠veres.
    - `'department_id'`: n√∫mero ID que identifica de manera √∫nica cada departamento de v√≠veres.
- `order_products.csv`: cada fila corresponde a un art√≠culo pedido en un pedido.
    - `'order_id'`: n√∫mero de ID que identifica de manera √∫nica cada pedido.
    - `'product_id'`: n√∫mero ID que identifica de manera √∫nica cada producto.
    - `'add_to_cart_order'`: el orden secuencial en el que se a√±adi√≥ cada art√≠culo en el carrito.
    - `'reordered'`: 0 si el cliente nunca ha pedido este producto antes, 1 si lo ha pedido.
- `aisles.csv`
    - `'aisle_id'`: n√∫mero ID que identifica de manera √∫nica cada categor√≠a de pasillo de v√≠veres.
    - `'aisle'`: nombre del pasillo.
- `departments.csv`
    - `'department_id'`: n√∫mero ID que identifica de manera √∫nica cada departamento de v√≠veres.
    - `'department'`: nombre del departamento.

````python
order_products_con_usuarios = order_products_df.merge(
    orders_df[['order_id', 'user_id']], 
    on='order_id', 
    how='left'
)

totales_usuario = order_products_con_usuarios.groupby('user_id').size().rename('total_productos')
reordenes_usuario = (
    order_products_con_usuarios[order_products_con_usuarios['reordered'] == 1]
    .groupby('user_id').size().rename('num_reordenes')
)

proporcion_reorden = pd.concat([totales_usuario, reordenes_usuario], axis=1).fillna(0)
proporcion_reorden['tasa_reorden'] = (
    proporcion_reorden['num_reordenes'] / proporcion_reorden['total_productos']
)

print("Estad√≠sticas de tasa de reorden por usuario:")
print(proporcion_reorden['tasa_reorden'].describe())

plt.figure(figsize=(12, 6))
plt.hist(proporcion_reorden['tasa_reorden'], bins=50, alpha=0.7,
         color='lightgreen', edgecolor='black')
plt.xlabel('Proporci√≥n de productos reordenados por usuario')
plt.ylabel('N√∫mero de usuarios')
plt.title('Distribuci√≥n de la proporci√≥n de productos reordenados por usuario')
plt.grid(axis='y', alpha=0.3)
plt.show()

print("\nEjemplos de comportamiento de usuarios:")
print(f"Usuarios que nunca reordenan: {(proporcion_reorden['tasa_reorden'] == 0).sum()}")
print(f"Usuarios que siempre reordenan: {(proporcion_reorden['tasa_reorden'] == 1).sum()}")
print(f"Tasa promedio de reorden: {proporcion_reorden['tasa_reorden'].mean():.3f}")
````
